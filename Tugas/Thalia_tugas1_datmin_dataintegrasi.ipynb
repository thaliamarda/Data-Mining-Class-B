# Data Integration Polusi Udara di Jakarta pada Desember 2018

```python
import pandas as pd
import numpy as np
```

```python
#membaca data
polusi = pd.read_csv ('polusi.csv')
indeks =pd.read_csv('indeks.csv')
polusi
```

```python
indeks
```

```python
#menggabungkan data
data = pd.merge(polusi, indeks, on=['Hari Ke','stasiun'])
data 
```

```python
#menghilangkan kolom
data1= data.drop(['critical'], axis=1)
data1
```

```python
#menghilangkan baris data yang mempunyai missing value
data2=data.dropna()
data2
```

```python
datacon= data2.iloc[:,2:7]
datacon
```

```python
#menghitung statistika deskriptif
statdes=datacon.describe()
statdes
```

```python
#mengganti missing value dengan mean data
ratapm=datacon['pm10'].mean()
data1['pm10']=data1['pm10'].fillna(ratapm)

rataso=datacon['so2'].mean()
data1['so2']=data1['so2'].fillna(rataso)

rataco=datacon['co'].mean()
data1['co']=data1['co'].fillna(rataco)

ratao=datacon['o3'].mean()
data1['o3']=data1['o3'].fillna(ratao)

ratano=datacon['no2'].mean()
data1['no2']=data1['no2'].fillna(ratano)

data1
```

```python
data3=data1.dropna()
data3
```

```python
data3["Total"] =(data3["pm10"]+data3["so2"]+data3["co"]+data3["o3"]+data3["no2"])
data3["Rata-Rata"] = data3["Total"]/5
data3
```
