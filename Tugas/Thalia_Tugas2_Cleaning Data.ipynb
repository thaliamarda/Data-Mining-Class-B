# Missing Value

```python
np.sum(data.isnull())
```

```python
data.isnull().sum().sum()
```

```python
#mengisi missing value pada setiap variabel
data['normalized-loses'] = data['normalized-loses'].fillna((data['normalized-loses'].mean())) 
data['num-of-doors'] = data.fillna(data['num-of-doors'].value_counts().index[0])
data['bore'] = data['bore'].fillna((data['bore'].mean()))
data['stroke'] = data['stroke'].fillna((data['stroke'].mean()))
data['horsepower'] = data['horsepower'].fillna((data['horsepower'].mean()))
data['peak-rpm'] = data['peak-rpm'].fillna((data['peak-rpm'].mean()))
data['price'] = data['price'].fillna((data['price'].mean()))
data.isnull().any()
```

# Encode Labels

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data['fuel-type'] = LabelEncoder().fit_transform(data['fuel-type'])
data['aspiration'] = LabelEncoder().fit_transform(data['aspiration'])
data['body-style'] = LabelEncoder().fit_transform(data['body-style'])
data['drive-wheels'] = LabelEncoder().fit_transform(data['drive-wheels'])
data['engine-location'] = LabelEncoder().fit_transform(data['engine-location'])
data.head()
```

# Standardize

```python
from sklearn.preprocessing import StandardScaler
```

```python
data['normalized-loses'] = StandardScaler().fit(data[['normalized-loses']]).transform(data[['normalized-loses']])
data['wheel-base'] = StandardScaler().fit(data[['wheel-base']]).transform(data[['wheel-base']])
data['bore'] = StandardScaler().fit(data[['bore']]).transform(data[['bore']])
data['stroke'] = StandardScaler().fit(data[['stroke']]).transform(data[['stroke']])
data['horsepower'] = StandardScaler().fit(data[['horsepower']]).transform(data[['horsepower']])
data['price'] = StandardScaler().fit(data[['price']]).transform(data[['price']])
data.head()
```

# Normalize

```python
from sklearn.preprocessing import Normalizer 
```

```python
data['normalized-loses'] = Normalizer().fit(data[['normalized-loses']]).transform(data[['normalized-loses']])
data['wheel-base'] = Normalizer().fit(data[['wheel-base']]).transform(data[['wheel-base']])
data['bore'] = Normalizer().fit(data[['bore']]).transform(data[['bore']])
data['stroke'] = Normalizer().fit(data[['stroke']]).transform(data[['stroke']])
data['horsepower'] = Normalizer().fit(data[['horsepower']]).transform(data[['horsepower']])
data['price'] = Normalizer().fit(data[['price']]).transform(data[['price']])
data.head()
```
