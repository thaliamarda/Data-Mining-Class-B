# Data Integration

```python
import pandas as pd
import numpy as np
```

```python
## Merging
tiket = pd.DataFrame({'kode': ['DGI01', 'ICL01', 'DGI02', 'ICL02'],
                     'Penerbangan': ['Domestik', 'Internasional', 'Domestik', 'Internasional'], 
                     'Maskapai': ['Garuda Indonesia', 'Citilink', 'Garuda Indonesia', 'Citilink']})

penumpang = pd.DataFrame({'kode': ['DGI01', 'ICL01', 'DGI02', 'ICL02'],
                      'JK': ['L', 'P', 'L', 'P'],
                      'Usia': ['45', '20', '36', '54']})

result1 = pd.merge(tiket, penumpang, on='kode')
result1
```

```python
## Merging different key id
tiket = pd.DataFrame({'kode': ['DGI01', 'ICL01', 'DGI02', 'ICL02', 'ICL03','ICL04'],
                     'Penerbangan': ['Domestik', 'Internasional', 'Domestik', 'Internasional','Internasional','Internasional'], 
                     'Maskapai': ['Garuda Indonesia', 'Citilink', 'Garuda Indonesia', 'Citilink', 'Citilink','Citilink']})

penumpang = pd.DataFrame({'kode': ['DGI01', 'ICL01', 'DGI02', 'ICL02'],
                      'JK': ['L', 'P', 'L', 'P'],
                      'Usia': ['45', '20', '36', '54']})

result2 = pd.merge(tiket, penumpang, on='kode')
result2
```

```python
# Ordering variable/column
result1 = result1[['kode', 'Penerbangan','Maskapai', 'JK', 'Usia']]
result1
```

```python
# Multiple key
tiket = pd.DataFrame({'kode': ['DGI01', 'ICL01', 'DGI02', 'ICL02'],
                     'Penerbangan': ['Domestik', 'Internasional', 'Domestik', 'Internasional'], 
                     'Maskapai': ['Garuda Indonesia', 'Citilink', 'Garuda Indonesia', 'Citilink'],
                     'Kelas': ['Bisnis', 'Bisnis', 'Ekonomi', 'Ekonomi']})

penumpang = pd.DataFrame({'kode': ['DGI01', 'ICL01', 'DGI02', 'ICL02'],
                      'JK': ['L', 'P', 'L', 'P'],
                      'Usia': ['45', '20', '36', '54'],
                           'Kelas': ['Bisnis', 'Eksekutif', 'Ekonomi', 'Bisnis']})

result3 = pd.merge(tiket, penumpang, on=['kode', 'Kelas'])
result3
```

```python
# Merge methods (default how="inner")
result_left = pd.merge(tiket, penumpang,how='left', on=['kode', 'Kelas'])
result_left
```

```python
# Merge methods (default how="inner")
result_right = pd.merge(tiket, penumpang,how='right', on=['kode', 'Kelas'])
result_right
```

```python
result_outer = pd.merge(tiket, penumpang,how='outer', on=['kode', 'Kelas'])
result_outer
```

```python
result_inner = pd.merge(tiket, penumpang,how='inner', on=['kode', 'Kelas'])
result_inner
```

```python
## Joining 
tiket = pd.DataFrame({'Penerbangan': ['Domestik', 'Internasional', 'Domestik', 'Internasional'], 
                     'Maskapai': ['Garuda Indonesia', 'Citilink', 'Garuda Indonesia', 'Citilink']},
                    index=['DGI01', 'ICL01', 'DGI02', 'ICL02'])

penumpang = pd.DataFrame({'JK': ['L', 'P', 'L', 'P'],
                      'Usia': ['45', '20', '36', '54']},
                        index=['DGI01', 'ICL01', 'ICL02', 'ICL03'])

biodata = pd.DataFrame({'JK': ['P', 'P', 'L'], 
                      'Usia': ['19', '21', '18']},
                    index=['id0', 'id1', 'id2'])
nilai = pd.DataFrame({'Kelas': ['Kalkulus', 'ADK', 'MO'],
                      'IPK': ['3.94', '3.34', '3.75']},
                    index=['id0', 'id2', 'id3'])
result = penumpang.join(tiket) 
result
```

```python
result = tiket.join(penumpang, how='outer')
result
```

```python
result = tiket.join(penumpang, how='inner') 
result
```

```python
result = pd.merge(tiket, penumpang, left_index=True, right_index=True, how='outer')
result
```

```python
result = pd.merge(tiket, penumpang, left_index=True, right_index=True, how='inner')
result
```

```python
tiket = pd.DataFrame({'kode': ['DGI01', 'ICL01', 'DGI02', 'ICL02'],
                     'Penerbangan': ['Domestik', 'Internasional', 'Domestik', 'Internasional'], 
                     'Maskapai': ['Garuda Indonesia', 'Citilink', 'Garuda Indonesia', 'Citilink']})

penumpang = pd.DataFrame({'JK': ['L', 'P', 'L'],
                      'Usia': ['45', '20', '36']},
                        index= ['DGI01', 'ICL01', 'DGI02'])
result = tiket.join(penumpang, on='kode')
result
```

```python
## Concatenating
data1 = pd.DataFrame({'A': ['A10', 'A11','A12'],
                'B': ['B10', 'B11','B12']})
data2 = pd.DataFrame({'A': ['A20', 'A21','A22'],
                'B': ['B20', 'B21','B22']})
data3 = pd.DataFrame({'A': ['A30', 'A31','A32'],
                'B': ['B30', 'B31','B32']})

data = [data1, data2, data3]
result = pd.concat(data)
result
```

```python
# Add category
result = pd.concat(data, keys=['X', 'Y', 'Z'])
result
```

```python
# Default join=outer (join all rows)
data4 = pd.DataFrame({'B': ['B2', 'B3', 'B4'],
                    'D': ['D2', 'D3', 'D4'],
                    'F': ['F2', 'F3', 'F4']},
                    index=[2, 3, 5])

result = pd.concat([data1, data4], axis=1)
result
```

```python
result = pd.concat([data1, data4], axis=1, join='inner')
result
```

```python
result = pd.concat([data1, data4], axis=1, join_axes=[data1.index])
result
```

```python
result = data1.append(data2)
result
```

```python
result = data1.append(data4)
result
```

```python
result = data1.append([data2, data3])
result
```

```python
result = pd.concat([data1, data2], ignore_index=True)
result
```

```python
result = data1.append(data4, ignore_index=True)
result
```

```python
datan = pd.Series(['X0', 'X1', 'X2', 'X3'], index=['A', 'B', 'C', 'D'])
result = data2.append(datan, ignore_index=True)
result
```
