```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
data=pd.read_csv('online_retail.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
np.sum(data.isnull())
```

```python
for i, j in data[data['Description'].isnull()].iterrows(): 
    data['Description'][i]="Code-"+str(j['StockCode'])
```

```python
a = data.groupby('InvoiceNo')
basket=[]
for i,j in a:
    basket.append(list(j['Description'].map(str)))
```

```python
data=pd.DataFrame(basket)
data.head()
```

```python
association_rules=apriori(basket,min_support=0.0087, 
                          min_confidence=0.18, 
                          min_lift=3, min_length=2) 
association_result=list(association_rules)
```

```python
association_result[0]
```

```python
for item in association_result:
    pair = item[0]
    items = [x for x in pair]
    print("rule: " + items [0] + " ->" + items[1])

    #second index of the inner list
    print("support: " + str(item[1]))

    #third index of the list located at 10th
    #of the third index of the inner list

    print("confidence: " + str(item[2][0][2]))
    print("lift: " + str(item[2][0][3]))
    print("=======================================")
```
